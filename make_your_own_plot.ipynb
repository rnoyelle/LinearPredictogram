```python
label1_cortex = []
for area in label1 :
    label1_cortex.append(dico_area_to_cortex[area])
    
label2_cortex = []
for area in label2 :
    label2_cortex.append(dico_area_to_cortex[area])
```

```python
target_cortex1 = 'Visual'
target_cortex2 = 'Prefrontal'

ind1 = (label1_cortex == target_cortex1)
ind2 = (label2_cortex == target_cortex2)
```

```python
FC_subset = FC[ind1,:]
FC_subset = FC[:,ind2]
label1_subset = label1[ind1]
label2_subset = label2[ind2]
```

```python
fig, ax = plt.subplots(figsize=(15,15))
threshold = 0.05 # I will add error bar file # therehold = error bar file or max(error bar file)
#fig, ax = plt.subplots()
im = ax.imshow(FC_subset, vmin=0)#, vmax=1.0)# cmap='jet')
# make the size of the colorbar the same as the picture 
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.10)
fig.colorbar(im, cax=cax)


# We want to show all ticks...
ax.set_xticks(np.arange(len(label2_subset)))
ax.set_yticks(np.arange(len(label1_subset)))
# ... and label them with the respective list entries
ax.set_xticklabels(label2_subset)
ax.set_yticklabels(label1_subset)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

#Loop over data dimensions and create text annotations.
for i in range(len(label1_subset)):
    for j in range(len(label2_subset)):
         if FC_subset[i, j] > threshold : # if r2 is bigger than 0+ error bar plot r2 score 
            # plot round number 
            # and with black or white according to the color of the pixel

            if round(FC_subset[i, j],3) < 0.8* FC_subset.max() :

                text = ax.text(j, i, round(FC_subset[i, j],1),
                               ha="center", va="center", color="w")
            else:
                text = ax.text(j, i, round(FC_subset[i, j],1),
                               ha="center", va="center", color="b")

plt.show()
```
